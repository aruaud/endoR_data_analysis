# Plots

```r
fnames <- list.files('./tmp/', pattern = 'plot', full.names = TRUE)
```

```r
plotz <- list()
i <- 1
for (f in fnames){
    plotz[[i]] <- readRDS(f)
    i <- i+1
}
```

```r
fnames
```

```r
plotz[[1]] <- plotz[[1]] + labs(title = 'r = 0.05, n = 1000, B = 10') + 
        theme(legend.position = 'bottom', legend.text = element_text(size = rel(1.2))
              , legend.title = element_text(size = rel(1.2)) )
plotz[[2]] <- plotz[[2]] + labs(title = expression('r = 0.05, n = 1000, '*alpha*' = 1'))+ 
        theme(legend.position = 'bottom', legend.text = element_text(size = rel(1.2))
              , legend.title = element_text(size = rel(1.2)) )
plotz[[3]] <- plotz[[3]] + labs(title = expression('r = 0.05, B = 10, '*alpha*' = 1'))+ 
        theme(legend.position = 'bottom', legend.text = element_text(size = rel(1.2))
              , legend.title = element_text(size = rel(1.2)) )
```

```r
p2 <- ggarrange(plotlist = plotz[1:3], labels = c('D', 'E', 'F'), ncol = 3, nrow = 1)
```

```r
p_all <- ggarrange(plotz[[4]], p2, ncol = 1, nrow = 2, heights = c(0.8, 1))
```

```r
options(repr.plot.width=12, repr.plot.height=7)
p_all
```

```r
ggsave(p_all, filename = '../Figures/Toy100.png', device = 'png', width = 12, height = 7, dpi = 300)
```
