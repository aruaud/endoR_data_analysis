# Simulate data

```r
simu <- data.table()

for (j in 1:10){
    set.seed(j)
    set(simu, NULL, paste0('V',j), value = rnorm(10000, mean = 1, sd = 1))
}
```

```r
set(simu, NULL, 'V4', value = rnorm(10000, mean = 0, sd = 1))
set(simu, NULL, 'V6', value = rnorm(10000, mean = 0, sd = 1))
```

```r
set(simu, 1:2500, 'target', simu[[1]][ 1:2500 ]*simu[[2]][ 1:2500 ] )
set(simu, 1:2500, 'group', 'a')

set(simu, 2501:5000, 'target', simu[[3]][ 2501:5000 ])
set(simu, 2501:5000, 'group', 'b')

set(simu, 5001:7500, 'target', simu[[4]][5001:7500 ]+simu[[5]][ 5001:7500 ])
set(simu, 5001:7500, 'group', 'c')

set(simu, 7501:nrow(simu), 'target', simu[[6]][7501:nrow(simu)] - simu[[7]][7501:nrow(simu)])
set(simu, 7501:nrow(simu), 'group', 'd')
```

```r
set(simu, NULL, 'target_ctg', as.character(sign(simu[['target']])))
```

```r
simu %>% group_by(group) %>% summarise(nsamples = n()
                                       , variance = round(sd(target)^2, digits = 3)
                                       , posi = sum(target_ctg == 1)/n()
                                      )
```

```r
dummies <- dummyVars(target_ctg ~ ., data = simu )
dummies <- as.data.table(predict(dummies, newdata = simu ))
dummies <- select(dummies, -target)
```

```r
simu %>% count(group, target_ctg) 
```

# Model

```r
simu <- simu[, group:=as.factor(group)]
```

```r
varnames <- c(colnames(simu)[1:7], 'group')
```

```r
set.seed(0)
rf_class <- randomForest(y = as.factor(simu$target_ctg)
                         #, mtry = 8
                         #, maxnodes = 5
                         , x = dummies
                        )
rf_class
```

```r
tmp <- as.data.frame(rf_class$importance)
tmp$Feature <- factor(rownames(tmp), levels = rownames(tmp)[order(tmp$MeanDecreaseGini)])
```

```r
options(repr.plot.width=3, repr.plot.height=3)
ggplot(tmp, aes(x = Feature, y = MeanDecreaseGini)) + 
geom_col(fill = 'lightgray', width = .1) +
geom_point(size = 2, aes(color = MeanDecreaseGini)) + 
theme_classic() + 
coord_flip() +
theme(axis.text.x = element_text(angle = 30, hjust = 1, vjust = 1)) +
scale_color_gradient2(low = 'lightgrey', high = "#D55E00", mid = "#E69F00", midpoint = 400, guide = FALSE)
```

# run

```r
test <- model2RE_dev(model = rf_class, model_type = 'rf', dummy_var = 'group'
                     ,data = dummies
                     ,target = as.factor(simu$target_ctg)
                     ,classPos = '1'
                     ,transform_decisions = "before"
                     ,discretize = TRUE, K = 3
                     ,prune = F
                     ,filter = F
                     ,in_parallel = TRUE, n_cores = 7
                    )
```

```r
options(repr.plot.width=8, repr.plot.height=5.5)
plotNetwork(edges = test$edges 
            , nodes = test$nodes
            , text_size = 4 
            #, n_edges = 15
            , hide_nodes = -1
            , layout = 'fr') +
labs(title = 'All edges')#+
#scale_edge_alpha(range = c(1, 1))
```

```r
options(repr.plot.width=8, repr.plot.height=5.5)
plotNetwork(edges = test$edges#, edges_value = 'wParticipation'
            , nodes = test$nodes
            , text_size = 4 
            , n_edges = 17
            , hide_nodes = -1
            , layout = 'fr') + 
theme(legend.position = "right"
     , legend.box = "vertical") + 
guides(size = guide_legend(ncol = 2, byrow = T)) +
labs(title = 'Best 17 edges')+
scale_edge_alpha(range = c(1, 1))
```

# Pruning

```r
test_p <- model2RE_dev(model = rf_class, model_type = 'rf', dummy_var = 'group'
                     ,data = dummies
                     ,target = as.factor(simu$target_ctg)
                     ,classPos = '1'
                     ,transform_decisions = "before"
                     ,discretize = TRUE, K = 2
                     ,prune = TRUE, maxDecay = 0.05, typeDecay = 2
                     ,filter = TRUE, min_imp = 1
                     ,in_parallel = TRUE, n_cores = 7
                    )
```

```r
test_p$rules_imp %>% nrow
```
