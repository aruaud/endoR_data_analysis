```r
model2RE_cluster <- function(partition){
    library(data.table)
    res <- model2RE_dev(data = data[partition, ], target = target[partition]
                      , exec = exec
                      , transform_decisions = "before"
                      , classPos = classPos
                      , prune = prune, maxDecay = maxDecay, typeDecay = typeDecay 
                      , discretize = discretize, K = K
                      , filter = filter, min_imp = min_imp, ntest = ntest
                      , in_parallel = in_parallel, n_cores = n_cores )
    res <- list('rules' = res[[length(res)-4]], 'nodes_agg' = res$nodes_agg, 'edges_agg' = res$edges_agg)
    return(res)
}
```

# Simulate data

```r
nr <- 5000
```

```r
simu <- data.table(tmp = 1:nr)

for (j in 1:12){
    set.seed(j)
    set(simu, NULL, paste0('V',j), value = rnorm(nr, mean = 1, sd = 1))
}
simu <- simu[,tmp:=NULL]
set(simu, NULL, 'V3', value = rnorm(nr, mean = 0.5, sd = 1))
set(simu, NULL, 'V4', value = rnorm(nr, mean = 0, sd = 1))
set(simu, NULL, 'V6', value = rnorm(nr, mean = 0, sd = 1))

set(simu, 1:(0.25*nr), 'target', simu[[1]][ 1:(0.25*nr) ]*simu[[2]][ 1:(0.25*nr) ] )
set(simu, 1:(0.25*nr), 'group', 'a')

set(simu, (0.25*nr+1):(0.5*nr), 'target', simu[[3]][ (0.25*nr+1):(0.5*nr) ])
set(simu, (0.25*nr+1):(0.5*nr), 'group', 'b')

set(simu, (0.5*nr+1):(0.75*nr), 'target', simu[[4]][(0.5*nr+1):(0.75*nr) ]+simu[[5]][ (0.5*nr+1):(0.75*nr) ])
set(simu, (0.5*nr+1):(0.75*nr), 'group', 'c')

set(simu, (0.75*nr+1):nr, 'target', simu[[6]][(0.75*nr+1):nr] - simu[[7]][(0.75*nr+1):nr])
set(simu, (0.75*nr+1):nr, 'group', 'd')
```

```r
brnounou <- rbinom(n = nr, size = 1,prob = 0.2)
allgroups <- c('a', 'b', 'c', 'd')
for (i in which(brnounou == 1)){
    set.seed(i)
    simu <- simu[i, group := sample(x = allgroups[allgroups != simu[['group']][i]],size = 1)]
}
```

```r
set(simu, NULL, 'target_ctg', as.character(sign(simu[['target']])))
```

```r
dummies <- dummyVars(target_ctg ~ ., data = simu )
dummies <- as.data.table(predict(dummies, newdata = simu ))
dummies <- select(dummies, -target)
```

```r
simu %>% group_by(group) %>% summarise(nsamples = n()
                                       , variance = round(sd(target)^2, digits = 3)
                                       , posi = sum(target_ctg == 1)/n()
                                      )
```

# Model

```r
simu <- simu[, group:=as.factor(group)]
```

```r
set.seed(0)
rf_class <- randomForest(y = as.factor(simu$target_ctg), x = dummies)
rf_class
```

```r
acc <- (rf_class$confusion[1,1] + rf_class$confusion[2,2])/nr
cat('Accuracy = ', acc)
```

```r
tmp <- as.data.frame(rf_class$importance)
tmp$Feature <- factor(rownames(tmp), levels = rownames(tmp)[order(tmp$MeanDecreaseGini)])
```

```r
options(repr.plot.width=3, repr.plot.height=3)
ggplot(tmp, aes(x = Feature, y = MeanDecreaseGini)) + 
geom_col(fill = 'lightgray', width = .1) +
geom_point(size = 2, aes(color = MeanDecreaseGini)) + 
theme_classic() + 
coord_flip() +
theme(axis.text.x = element_text(angle = 30, hjust = 1, vjust = 1)) +
scale_color_gradient2(low = 'lightgrey', high = "#D55E00", mid = "#E69F00", midpoint = 400, guide = FALSE)
```

# Network accuracy

```r
accuracyEdges <- function(pred, truth, colN){
    # truth: list of vectors of length 2 with the true pairs of associated variables
    # pred: same but as predicted by the method
    all <- asplit(combn(colN, m = 2), MARGIN = 2)
    truth <- lapply(truth, sort)
    pred <- lapply(pred, sort)
    
    
    truth <- sapply(truth, function(x){paste(x, collapse = ' - ')})
    pred <- sapply(pred, function(x){paste(x, collapse = ' - ')})
    
    # those that should be and are = in both
    #tp <- length(which(sapply(pred, function(x, truth){x %in% truth}, truth = truth)))
    
    # those that should not be but are = in pred but not truth
    fp <- length(which(sapply(pred, function(x, truth){!(x %in% truth)}, truth = truth)))
    
    # those that should be but are not = in truth but not in pred
    fn <- length(which(sapply(truth, function(x, pred){!(x %in% pred)}, pred = pred)))
    
    #accuracy:
    acc <- 1-(fp+fn)/length(all)
    return(acc)
}
```

```r
true_edges <- readRDS('/ebio/abt3_projects/Methanogen_SCFA/Metagenomes_methanogen/RF_coocc/RF_coocc/tmp/true_edgesK3.RDS')
```

```r
alldata <- readRDS('/ebio/abt3_projects/Methanogen_SCFA/Metagenomes_methanogen/RF_coocc/RF_coocc/tmp/nv5p02.RDS')
```

```r
pred1 <- select(alldata$edges, c('x', 'y'))
pred1 <- asplit(as.matrix(pred1), MARGIN = 1)
cat('Accuracy network on a 1 subset = ')
accuracyEdges(pred1, truth = true_edges, colN = colnames(alldata$data_ctg))
```

```r
pred1_20 <- select((alldata$edges %>% arrange(desc(participation)))[1:20,], c('x', 'y'))
pred1_20 <- asplit(as.matrix(pred1_20), MARGIN = 1)
cat('Accuracy network on a 1 subset, best 20 edges = ')
accuracyEdges(pred1_20, truth = true_edges, colN = colnames(alldata$data_ctg))
```

```r
pred100 <- select(net100$edges, c('x', 'y'))
pred100 <- asplit(as.matrix(pred100), MARGIN = 1)
cat('Accuracy network on a 100 subset = ')
accuracyEdges(pred100, truth = true_edges, colN = colnames(alldata$data_ctg))
```
