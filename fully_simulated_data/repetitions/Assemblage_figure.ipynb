# Plots

```r
fnames <- list.files('./', pattern = '\\_plot', full.names = TRUE)
```

```r
fnames
```

```r
plotz <- list()
i <- 1
for (f in fnames){
    plotz[[i]] <- readRDS(f)
    i <- i+1
}
```

```r
plotz[[1]] <- plotz[[1]] + labs(title = expression('r = 0.05, n = 1000, B = 10, '*alpha*' varies')) + 
        theme(legend.position = 'bottom', legend.text = element_text(size = rel(1.2))
              , legend.title = element_text(size = rel(1.2)) )
plotz[[2]] <- plotz[[2]] + labs(title = expression('r = 0.05, n = 1000, B varies, '*alpha*' = 5'))+ 
        theme(legend.position = 'bottom', legend.text = element_text(size = rel(1.2))
              , legend.title = element_text(size = rel(1.2)) )
plotz[[3]] <- plotz[[3]] + labs(title = expression('r = 0.05, n varies, B = 10, '*alpha*' = 5'))+ 
        theme(legend.position = 'bottom'
              #, legend.direction = "horizontal"
              , legend.text = element_text(size = rel(1.2))
              , legend.title = element_text(size = rel(1.2)) )
```

```r
p2 <- ggarrange(plotlist = plotz[1:3], labels = c('G', 'H', 'I'), ncol = 3, nrow = 1)
```

```r
p_all <- ggarrange(plotz[[4]], plotz[[5]]
                   , p2, ncol = 1, nrow = 3, heights = c(0.8, 0.8, 1))
```

```r
options(repr.plot.width=12, repr.plot.height=10)
p_all
```

```r
ggsave(p_all, filename = 'Toy100.png', device = 'png', width = 12, height = 10, dpi = 300)
```
