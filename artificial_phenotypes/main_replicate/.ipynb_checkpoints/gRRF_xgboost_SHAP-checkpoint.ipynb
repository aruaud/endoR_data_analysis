# Functions

```r
getSingleInteractions <- function(data_long,data_int, x,y){
    data0 <- data.frame(x = rep(x, nrow(data_int))
                        , y = rep(y, nrow(data_int))
                        , shap_int = data_int[, x, y] )
    return(data0)
}
```

```r
getAllInteractions <- function(data_long,data_int){
    colN <- as.character(unique(data_long$variable))
    res <- list()
    k <- 1
    for (i in 1:(length(colN)-1)){
        for (j in (i+1):length(colN)){
            res[[k]] <- getSingleInteractions(data_long, data_int, x = colN[i],y = colN[j])
            k <- k+1
        }
    }
    
    res <- do.call(rbind, res)
    
    return(res)
}
```

# Make the target

```r
# draw random variables for predicting groups
set.seed(1209)
var_ix <- sample(which(colSums(X != 0) > nrow(X)/2), 9, replace = FALSE)
var_n <- colnames(X)[var_ix]
```

```r
var_n
```

```r
nr <- nrow(X)
ng <- floor(nr/4)
# make groups
X <- as.data.table(X)[,'group':=c(rep('a',ng), rep('b', ng), rep('c', ng), rep('d', nr-3*ng))]
```

```r
target <- data.frame('group'= X$group, 'tc'='1', stringsAsFactors = FALSE)
```

```r
# for each group, make target according to random drawn taxa of var_n
target$tc[target$group == 'a'] <- ifelse( X$s_Marvinbryantia_sp900066075[target$group == 'a'] > 0 
                                         & X$g_Alistipes_A[target$group == 'a'] > 0 
                                         , '1', '-1')
target$tc[target$group == 'b'] <- ifelse( X$f_Bacteroidaceae[target$group == 'b'] > 10^-(1) 
                                         & X$g_Dialister[target$group == 'b'] > 10^-(2.5)
                                         , '1', '-1')
target$tc[target$group == 'c'] <- ifelse((X$s_Oscillibacter_sp001916835[target$group == 'c'] > 0 
                                         & X$s_Bacteroides_clarus[target$group == 'c'] > 0)
                                         | X$s_Faecalibacterium_prausnitzii_G[target$group == 'c'] >10^-2
                                         , '1', '-1')
target$tc[target$group == 'd'] <- ifelse( X$s_Lawsonibacter_sp000177015[target$group == 'd'] <= 10^-3.4 
                                         & X$f_Anaerovoracaceae[target$group == 'd'] > 0
                                         , '1', '-1')
```

```r
# randomise group labels
groups <- c('a', 'b', 'c', 'd', 'e')
set.seed(0)
brnounou <- rbinom(n = length(X$group), size = 1,prob = 0.05)
for (i in 1:length(brnounou)){
    if (brnounou[i] == 1){
        set.seed(i)
        X$group[i] <- sample(groups[groups != X$group[i]], 1)
    }
}
```

# Model

```r
target_c <- as.factor(target$tc)
X <- X[, 'group':= as.factor(group)]
```

```r
# transform to dummy
dummies <- dummyVars(~ ., data = X )
dummies <- as.data.table(predict(dummies, newdata = X ))
```

```r
colnames(dummies) <- colnames(dummies) %>% str_replace_all(pattern = '\\.', replacement ='')
```

```r
gamma <- 0.45
```

```r
set.seed(0)
message('Feature selection')
RF <- RRF(dummies, flagReg=0, as.factor(target_c))
imp <-RF$importance[,"MeanDecreaseGini"]
impRF <- (imp - min(imp))/(max(imp) - min(imp))# normalization
coefReg <- (1-gamma) + gamma*impRF
GRRF <- RRF(dummies, as.factor(target_c), flagReg=1, coefReg=coefReg)
```

```r
# select data
to_keep <- colnames(dummies)[GRRF$feaSet]
X_fs <- select(dummies, all_of(to_keep))
```

```r
y <- ifelse(target_c == '1', 1, 0)
```

```r
set.seed(0)
xgb_fs <- xgboost::xgboost(data = as.matrix(X_fs), label = y, objective = 'binary:logistic', nrounds = 10)
```

# xgboost accuracy

Quick and dirty without the FS.. 

```r
set.seed(0)
trainIx <- createDataPartition(y = y, times = 10, p = .7, list = TRUE)
```

```r
# could use confusionMatrix()
getKappa <- function(x){
    p0 <- (x[1,1] + x[2,2]) / sum(x)
    py <- (x[1,1] + x[1,2])*(x[1,1] + x[2,1]) / sum(x)^2
    pn <- (x[2,1] + x[2,2])*(x[1,2] + x[2,2]) / sum(x)^2
    pe <- py+pn
    
    k <- 1 - (1-p0)/(1-pe)
    return(k)
}
```

```r
getSingleXGBkappa <- function(ix, model, data, target){
    xgb_pred <- predict(object = model, data[ix, ])
    xgb_pred <- as.matrix(table(round(xgb_pred, digits = 0), target[ix]))
    return(getKappa(xgb_pred))
}
```

```r
kappas <- sapply(trainIx, getSingleXGBkappa, model = xgb_fs, data = as.matrix(X_fs), target = y)
```

```r
range01 <- function(x){ 
    # change the zeros to get nicer log10 plots
    mini <- min(x[x > 0]) 
    mini <- mini - mini/10
    x[x == 0] <- mini
        
    x <- log10(x)
    (x - min(x))/(max(x) - min(x))
}
```

```r
getPlotsShaps <- function(x, shap_long, shap_int){
    if (str_detect(x, pattern =':')){
        y <- str_extract(x, pattern = '(?<=:).*')
        x <- str_extract(x, pattern = '.*(?=:)')
        if (str_detect(x, pattern = 'group')){tmp<-x;x<-y;y<-tmp}
        p <- shap.plot.dependence(data_long = shap_long,data_int = shap_int
                         , x= x, y = y, color_feature = y, smooth = FALSE)
        
    } else { p <- shap.plot.dependence(data_long = shap_long, x= x, color_feature=x, smooth = FALSE) }
    
    return(p)
}
```

```r
to_plot <- as.character(subset(shap_allvars, !str_detect(variable, pattern = ':')) %>% arrange(-mean_value) %>% 
                    head(5) %>% select(variable) %>% unlist)
to_plot <- c(to_plot, as.character(subset(shap_allvars, str_detect(variable, pattern = ':')) %>% arrange(-mean_value) %>% 
                    head(9) %>% select(variable) %>% unlist))
p_list <- lapply( as.character(to_plot), getPlotsShaps, shap_long=shap_long01, shap_int=shap_int)
```

```r
# additional homemade functions to change the labels of plots so they fit
shortenNames <- function(x){
    x %>%  str_replace(pattern = '(?<=s\\_[:upper:])[:lower:]+(?=\\_)', replacement = '') %>% 
           str_replace(pattern = '(?<=(^| ))[:lower:]\\_', replacement = '') %>% 
           str_replace(pattern = '\\n\\(Feature value\\)', replacement = '') %>% 
           str_replace(pattern = ' value(s?) for', replacement = '')
}
shortLabs <- function(x){
    x$labels <- lapply(x$labels, shortenNames)
    return(x)
}
```

```r
p_list <- lapply(p_list, shortLabs)
```
